```python
import sklearn.datasets as datasets
import pandas as pd
from pathlib import Path

iris_data = datasets.load_iris()
iris_data

# export iris_data to a csv file along with feature names and target names  

df = pd.DataFrame(data=iris_data.data, columns=iris_data.feature_names)
df['target'] = iris_data.target

# save the dataframe to a csv file in the raw_data folde
out_dir = Path('../raw')
df.to_csv(out_dir / 'iris_data.csv', index=False)
```
