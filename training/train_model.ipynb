```python
import sklearn.model_selection  as train_test_split
import pandas as pd
from sklearn.linear_model import LinearRegression   
```

```python
# load data from file
data = pd.read_csv('../data/raw/iris_data.csv')
data.head()

X_train, X_test, y_train, y_test = train_test_split.train_test_split(data[data.columns[:-1]], data['target'], test_size=0.2, random_state=42)
model = LinearRegression()  
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
y_pred

# save model to file
import joblib
joblib.dump(model, '../model/iris_data_model.joblib')
```
